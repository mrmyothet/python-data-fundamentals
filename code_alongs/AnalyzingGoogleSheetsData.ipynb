### Analyzing Google Sheets Data in Python

```python
import gspread 
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
```

```python
# Define scope
scope = ["https://www.googleapis.com/auth/spreadsheets",
         "https://www.googleapis.com/auth/drive"]

# Load credentials
creds = ServiceAccountCredentials.from_json_keyfile_name("credentials.json", scope)
client = gspread.authorize(creds)

# spreadsheet = client.open_by_url("https://docs.google.com/spreadsheets/d/1kcgqEsTRh3M5hn3i23ZkL-auqJJdBouXzG4_6rC_jq8/edit")
# sheet = spreadsheet.worksheet("ga_sessions")

# Open a sheet by name
sheet = client.open("ga_sample_data_codealong")

# Open worksheet (tab) by name
sheet = spreadsheet.worksheet("ga_sessions")

# Read data
data = sheet.get_all_records()
print(data)
```
